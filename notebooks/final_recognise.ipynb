**with perspective transformation**

```python
import cv2
import numpy as np
import os
from glob import glob

# Set paths
image_dir = '/content/runs/obb/predict'
label_dir = '/content/runs/obb/predict/labels'  # Labels are here
output_dir = os.path.join(image_dir, 'cropped')
os.makedirs(output_dir, exist_ok=True)

# Resize constraints
min_width, min_height = 200, 100
optimal_min_width, optimal_min_height = 400, 200
optimal_max_width, optimal_max_height = 800, 400

def order_points(pts):
    rect = np.zeros((4, 2), dtype="float32")
    s = pts.sum(axis=1)
    rect[0] = pts[np.argmin(s)]  # top-left
    rect[2] = pts[np.argmax(s)]  # bottom-right
    diff = np.diff(pts, axis=1)
    rect[1] = pts[np.argmin(diff)]  # top-right
    rect[3] = pts[np.argmax(diff)]  # bottom-left
    return rect

# Loop over each image
image_paths = glob(os.path.join(image_dir, '*.jpg'))
for image_path in image_paths:
    img = cv2.imread(image_path)
    if img is None:
        print(f"âŒ Could not load image {image_path}")
        continue
    h, w = img.shape[:2]

    # Corresponding label file in the label directory
    image_basename = os.path.basename(image_path).rsplit('.', 1)[0]
    label_path = os.path.join(label_dir, image_basename + '.txt')
    if not os.path.exists(label_path):
        print(f"âš ï¸ Label file not found for {image_path}")
        continue

    with open(label_path, 'r') as f:
        lines = f.readlines()

    for idx, line in enumerate(lines):
        parts = line.strip().split()
        if len(parts) != 9:
            print(f"âš ï¸ Skipping invalid label in {label_path}: {line.strip()}")
            continue

        coords = np.array(parts[1:], dtype=np.float32).reshape(4, 2)
        coords_pixel = coords * np.array([[w, h]])
        ordered_pts = order_points(coords_pixel)

        widthA = np.linalg.norm(ordered_pts[2] - ordered_pts[3])
        widthB = np.linalg.norm(ordered_pts[1] - ordered_pts[0])
        maxWidth = int(max(widthA, widthB))

        heightA = np.linalg.norm(ordered_pts[1] - ordered_pts[2])
        heightB = np.linalg.norm(ordered_pts[0] - ordered_pts[3])
        maxHeight = int(max(heightA, heightB))

        dst_pts = np.array([
            [0, 0],
            [maxWidth-1, 0],
            [maxWidth-1, maxHeight-1],
            [0, maxHeight-1]
        ], dtype="float32")

        M = cv2.getPerspectiveTransform(ordered_pts, dst_pts)
        warped = cv2.warpPerspective(img, M, (maxWidth, maxHeight))

        # Resize according to constraints
        if maxWidth < optimal_min_width or maxHeight < optimal_min_height:
            warped_resized = cv2.resize(warped, (min_width, min_height))
        elif maxWidth <= optimal_max_width and maxHeight <= optimal_max_height:
            warped_resized = cv2.resize(warped, (maxWidth, maxHeight))
        else:
            warped_resized = cv2.resize(warped, (optimal_max_width, optimal_max_height))

        output_file = os.path.join(output_dir, f"{image_basename}_crop_{idx}.jpg")
        cv2.imwrite(output_file, warped_resized)
        print(f"âœ… Saved {output_file}")
```

```python
# @title
from paddleocr import PaddleOCR
import os
import csv
from glob import glob

# Initialize OCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')

# Define directories
cropped_folder = '/content/runs/obb/predict/cropped'
output_csv = '/content/license_plates_final.csv'

# Get all image paths in cropped folder
image_paths = glob(os.path.join(cropped_folder, '*.jpg'))

# Create/open CSV file
with open(output_csv, mode='w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['image_name', 'license_plate_text', 'confidence'])  # header

    # Loop over each cropped image
    for img_path in image_paths:
        image_name = os.path.basename(img_path)
        result = ocr.ocr(img_path, cls=True)

        if result and result[0]:
            for line in result[0]:
                text = line[1][0]
                conf = line[1][1]
                writer.writerow([image_name, text, f"{conf:.2f}"])
                print(f"{image_name} â†’ {text} ({conf:.2f})")
        else:
            # No text detected
            writer.writerow([image_name, "None", "0.00"])
            print(f"{image_name} â†’ None (0.00)")

print(f"\nâœ… Results saved to {output_csv}")
```
